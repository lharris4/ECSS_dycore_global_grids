```python
import numpy as np
import matplotlib.pyplot as plt
from netCDF4 import Dataset
import xarray as xr
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter
import matplotlib.pyplot as plt
import matplotlib
import cartopy.crs as ccrs
import cartopy.feature as cfeature
%matplotlib inline
import os

# os.chdir('/lustre/f2/scratch/gfdl/Joseph.Mouallem/solo/test_solo/gridmovie')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(4, 4), 
                       subplot_kw={'projection': ccrs.Orthographic(central_longitude=-20.0, central_latitude=0)})

# ax.add_feature(cfeature.NaturalEarthFeature(
#     'physical', 'coastline', '50m',
# edgecolor='grey', facecolor='none', linewidth=0.5))
plt.title('Gnomonic Cubed-Sphere Grid',fontsize=14)

loc="/home/lmh/research/PySandbox/global_grid_plots/gnomonic_cubic_grid"

ext=''
color='b'
linestyle='solid'

loc=loc+ext 
print(loc)
linewidth=0.5
plot_edges = False
for n in range(1,7):
   
     # file_n = loc+"/grid_spec.tile"+str(i)+".nc"
     # print("Read from tile"+str(i))
     # fn = Dataset(file_n, 'r')

     # lat_n = fn.variables['grid_lon'][:]
     # lon_n = fn.variables['grid_lat'][:]
     
     file_n = loc+"/horizontal_grid.tile"+str(n)+".nc"
     print("Read from tile"+str(n))
     fn = Dataset(file_n, 'r')

     lat_n = fn.variables['x'][:]
     lon_n = fn.variables['y'][:]
    
     lat_n1 = lat_n[0,:]
     lat_n2 = lat_n[:,0]
     lat_n3 = lat_n[:,-1]
     lat_n4 = lat_n[-1,:]

     lon_n1 = lon_n[0,:]
     lon_n2 = lon_n[:,0]
     lon_n3 = lon_n[:,-1]
     lon_n4 = lon_n[-1,:]

#      color = next(ax._get_lines.prop_cycler)['color'] #change color every iteration
     color='k'

     if plot_edges:
         ax.plot(lat_n1,lon_n1,color=color,transform=ccrs.PlateCarree(),linestyle=linestyle,linewidth=linewidth)
         ax.plot(lat_n2,lon_n2,color=color,transform=ccrs.PlateCarree(),linestyle=linestyle,linewidth=linewidth)
         ax.plot(lat_n3,lon_n3,color=color,transform=ccrs.PlateCarree(),linestyle=linestyle,linewidth=linewidth)
         ax.plot(lat_n4,lon_n4,color=color,transform=ccrs.PlateCarree(),linestyle=linestyle,linewidth=linewidth)
        
     dx = 1
     dy = dx
     nx,ny = np.shape(lat_n)
     idx_list = np.arange(0,nx,dx)
     idy_list = np.arange(0,ny,dy)
     # idx_list = np.arange(0,nx-1)
     # idy_list = np.arange(0,ny-1)
     for i in idx_list:
        lat_n1 = lat_n[i,:]
        lon_n1 = lon_n[i,:]
        ax.plot(lat_n1,lon_n1,color=color,linewidth=0.5,transform=ccrs.Geodetic())


     for i in idy_list:
        lat_n1 = lat_n[:,i]
        lon_n1 = lon_n[:,i]
        ax.plot(lat_n1,lon_n1,color=color,linewidth=0.5,transform=ccrs.Geodetic())

plt.tight_layout()
plt.savefig('gnomonic.pdf', bbox_inches='tight')
plt.show() 
```
